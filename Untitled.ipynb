```python
#importing libraries
import pandas as pd
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#uploading dataset
df1 = pd.read_csv(r'C:\Users\HP\Downloads\Housing Sales Datasets\HOUSE_PRICES_LESS.csv')
df2 = pd.read_csv(r'C:\Users\HP\Downloads\Housing Sales Datasets\housing_FULL.csv')
```

```python
#checking null values
df1.isnull().sum()
```

```python
df2.isnull().sum()
```

```python
#removing null values
df_1 = df1.dropna(subset=["Price"])
```

```python
df_2 = df2.dropna(subset=["Price","CouncilArea"])
```

```python
#merging both files
df_1 = df_1[['Suburb', 'Address', 'Rooms', 'Type','Price', 'Method','SellerG', 'Date', 'Regionname','Distance', 'CouncilArea']]
df_2 = df_2[['Suburb', 'Address', 'Rooms', 'Type', 'Price', 'Method','SellerG', 'Date', 'Regionname','Distance', 'CouncilArea']]
```

```python
concatenated = pd.concat([df_1, df_2])
```

```python
#removing data with method "Property passed in(PI)" & rows with "Na"
concatenated = concatenated.drop(concatenated[concatenated['Method'] == 'PI'].index)
concatenated = concatenated.drop(concatenated[concatenated['Distance'] == '#N/A'].index)
concatenated = concatenated.drop(concatenated[concatenated['CouncilArea'] == '#N/A'].index)
concatenated = concatenated.drop(concatenated[concatenated['Regionname'] == '#N/A'].index)
```

```python
# lets find the sales trend year wise
concatenated['Date'] = pd.to_datetime(concatenated['Date'])
concatenated['Year'] = concatenated['Date'].dt.year
sales_per_year = concatenated.groupby('Year').size()

# create a bar chart of the sales per year
sales_per_year.plot(kind='bar')
plt.xlabel('Year')
plt.ylabel('Number of Sales')
plt.title('Sales per Year')
plt.show()
```

```python
#sales per month for all years
concatenated['Month'] = concatenated['Date'].dt.month
concatenated['Year'] = concatenated['Date'].dt.year
sales_per_month = concatenated.groupby(['Year', 'Month']).size()

# plot the sales per month for all years
fig, ax = plt.subplots(figsize=(10, 6))
sales_per_month.plot(kind='line', ax=ax)
ax.set_xlabel('Year,Month')
ax.set_ylabel('Number of Sales')
ax.set_title('Sales per Month for All Years')
plt.show()
```

```python
num_bins = 60

# Create a histogram to show the distribution of distances
plt.hist(concatenated['Distance'], bins=num_bins)
plt.xlabel('Distance')
plt.ylabel('Sales')
plt.title('sales as per Distance from CBD')
plt.show()
```

```python
# count the number of sales per month for 2018
sales_2018 = concatenated[concatenated['Year'] == 2018]
sales_per_month_2018 = sales_2018.groupby(['Month']).size()

# plot the sales per month for 2018
fig, ax = plt.subplots(figsize=(10, 6))
sales_per_month_2018.plot(kind='line',ax=ax)
ax.set_xlabel('Month')
ax.set_ylabel('Number of Sales')
ax.set_title('Sales per Month in 2018')
plt.show()
```

```python
#Top 10 Suburbs with most Sales
sales_per_region = concatenated.groupby(['Suburb']).size()
top_10_regions = sales_per_region.sort_values(ascending=False).head(10)
# plot top 10 suburbs with most sales
fig, ax = plt.subplots(figsize=(12, 8))
top_10_regions.plot(kind='barh', ax=ax)
ax.set_xlabel('Sales count')
ax.set_ylabel('Suburb')
ax.set_title('Top 10 Suburbs with most Sales')
plt.show()
```

```python
#Cheapest suburbs acc to avg price
suburb_avg_price = concatenated.groupby('Suburb')['Price'].mean().sort_values(ascending=True).head(10)

# plot the average price for each suburb
fig, ax = plt.subplots(figsize=(12, 8))
suburb_avg_price.plot(kind='barh', ax=ax)
ax.set_xlabel('Average Price')
ax.set_ylabel('Suburb')
ax.set_title('Suburb with lowest avg price')
plt.show()
```
